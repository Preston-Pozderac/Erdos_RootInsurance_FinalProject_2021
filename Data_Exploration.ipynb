```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Rootdf = pd.read_csv('Root_Insurance_data_Edited.csv')
```

### Adding Interaction Variables

```python
Rootdf['Mult_Vehicles_Drivers'] = Rootdf.Num_Vehicles * Rootdf.Num_Drivers
print(Rootdf.Mult_Vehicles_Drivers.describe())
Rootdf['Add_Vehicles_Drivers'] = Rootdf.Num_Vehicles + Rootdf.Num_Drivers
print(Rootdf.Add_Vehicles_Drivers.describe())
```

## Describing Data Overall

```python
print(Rootdf.describe())
print(Rootdf.describe(include='object'))
```

## Breakdown of Insured, Marital, Num_Vehicles, Num_Drivers, and Interactions

```python
print('Insured Breakdown')
print(Rootdf.Insured.value_counts())
print(Rootdf.Insured.value_counts()/Rootdf.Insured.count()*100)
print()

print('Marital Status Breakdown')
print(Rootdf.Marital.value_counts())
print(Rootdf.Marital.value_counts()/Rootdf.Marital.count()*100)
print()

print('Number of Vehicles Breakdown')
print(Rootdf.Num_Vehicles.value_counts())
print(Rootdf.Num_Vehicles.value_counts()/Rootdf.Num_Vehicles.count()*100)
print()

print('Number of Drivers Breakdown')
print(Rootdf.Num_Drivers.value_counts())
print(Rootdf.Num_Drivers.value_counts()/Rootdf.Num_Drivers.count()*100)
print()

print('Number of Drivers times Number of Vehicles Breakdown')
print(Rootdf.Mult_Vehicles_Drivers.value_counts())
print(Rootdf.Mult_Vehicles_Drivers.value_counts()/Rootdf.Mult_Vehicles_Drivers.count()*100)
print()

print('Number of Drivers plus Number of Vehicles Breakdown')
print(Rootdf.Add_Vehicles_Drivers.value_counts())
print(Rootdf.Add_Vehicles_Drivers.value_counts()/Rootdf.Add_Vehicles_Drivers.count()*100)
print()
```
